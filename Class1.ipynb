```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation 

x=[]
y=[]

fig, ax = plt.subplots() 

ax.set_xlim(0,20)
ax.set_ylim(0, 20)
  
line, = ax.plot(0,0)

def animo(i):
  x.append(i* 10)
  y.append(i)

  line.set_xdata(x)
  line.set_ydata(y) 
  return line,

anim = FuncAnimation(fig, func= animo,  frames = 100, interval = 10) 
plt.show()
```
